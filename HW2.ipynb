```
#değişkenler
ogrenciIsimListesi=[["Ömer","Yazar"],["Kutay","Çözer"],["Aysuda","Bilir"],["Ezgi","Okur"],["Mert","Güder"]]

# notlar=[["araSınav",0],["final",0],["ödev",0]]

bilgiler={}


#fonksiyonlar
# import random as rnd

# def sayiUydur():
#   return rnd.randint(0, 100)



def karsilasTir(bilgiler):
  birinci=""
  enYuksekNot=0
  
  for i in bilgiler:
    if bilgiler[i][1]>enYuksekNot:
      enYuksekNot=bilgiler[i][1]
      birinci=i

  print(f"\n\nTebrikler {birinci} !!!\n\n\n")



def ortalamaHesapla(*args):
  sonuc=0
  
  for i in args:
    sonuc=sonuc+i

  return sonuc/len(args)




def notlariGir():
  notListesi=[["araSınav",0],["final",0],["ödev",0]]

  # x=sayiUydur()
  notListesi[0][1]=int(input(f"Ara Sınav Notu: "))
  # notlar[0][1]=x

  # x=sayiUydur()
  notListesi[1][1]=int(input(f"Final Notu: "))
  # notlar[1][1]=x

  # x=sayiUydur()
  notListesi[2][1]=int(input(f"Ödev Notu: "))
  # notlar[2][1]=x

  print("\n\n")

  return notListesi


def notlariAta(isim,soyisim):
  print(f"{isim} {soyisim} adlı öğrencimizin notlarını girmektesiniz.\n")

  notListesi = notlariGir()

  bilgiler[isim+" "+soyisim]=[notListesi,ortalamaHesapla(notListesi[0][1],notListesi[1][1],notListesi[2][1])]



def main():
  print("            _ _ _ÖĞRENCİ BİLGİ SİSTEMİ_ _ _   ")
  print("\n")
  while True:
    secim=input("Öğrencilerimizin notlarını girmeye başlamak için herhangi bir değer giriniz. Eğer çıkmak istiyorsanız için 'çık' veya 'ç' girin :")
    if secim=='çık' or secim=='ç':
      print("\n...  Sistemden çıkılmıştır  ...")
      break
    print("\n")

    for i in ogrenciIsimListesi:
      notlariAta(i[0],i[1])

    print("      Öğrencilerimize ait bilgiler aşağıdaki gibidir:")
    print("\n   İsim-Soyisim       Ara Sınav      Final           Ödev          Ortalama")
    print("  ---------------     -----------   ---------      --------      ------------")
    for i in bilgiler:
      print(f"*  {i}             {bilgiler[i][0][0][1]}          {bilgiler[i][0][1][1]}             {bilgiler[i][0][2][1]}           {bilgiler[i][1]}")

    karsilasTir(bilgiler)


main()  
```
