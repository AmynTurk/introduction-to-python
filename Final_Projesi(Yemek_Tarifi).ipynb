```
#global fonksiyonlar ve değişkenler
yemekListesi=["Pizza","Poğaça","Pişi"]


def tarifiGor(obj):
  choice=print("\n\n\n"+obj.yemekTuru + " türündeki ve " + obj.yemekAdi + " adındaki yemeğimizin tarifi aşağıdaki gibidir:")
  obj.tarifiGoster()

  print("Afiyet Olsun!\n\n\n\n")

def listele(elemanlar):
  for i in elemanlar:
    print("*"+elemanlar[i]+" "+i , end=" \n")




#SINIFLAR
#Dede Sınıf
class Yemekler:
  def __init__(self):
    self.yemekAdi=""
    self.yemekTuru=""
    self.malzemeler=dict()
  def malzemeListesiGoster(self,malzemeler):
    for i in malzemeler:
      print(malzemeler[i]+" "+i , end=", ")

#Baba Sınıf
class HamurIsi(Yemekler):
  def __init__(self):
    super().__init__()
    self.yemekTuru="Hamur İşi"
    self.hamurMalzemeleri=dict()
  def hamuruKar(self,malzemeListesi):
    print("\n")
    for i in malzemeListesi:
      print(malzemeListesi[i]+" "+i , end=", ")
    print("hepsini karıştırınız.\n")
  def hamuruYogur(self):
    print("Karıştırmış olduğunuz hamuru iyice yoğurunuz.\n")

#Oğul Sınıflar
class Acik(HamurIsi):
  def __init__(self):
    super().__init__()
  def hamuruYay(self):
    print("Yoğurulmuş hamuru tepsiye yayınız.\n")

class Kapali(HamurIsi):
  def __init__(self):
    super().__init__()
  def hamuruEsitOlarakBol(self):
    print("Yoğurulmuş hamuru, hazırlamak istediğiniz sayıya göre eşit olarak bölünüz.\n")

class Yagda(HamurIsi):
  def __init__(self):
    super().__init__()
  def yagdaKizart(self):
    print("Hamurları birbirlerine temas etmeyecek şekilde yağda kızarmaya bırakınız, kabarıp pembeleşince alınız.\n")

class Firinda(HamurIsi):
  def __init__(self):
    super().__init__()
    self.sosMalzemeleri=dict()
    self.icMalzemeleri=dict()
  def sosHazirla(self,malzemeListesi):
    self.sosMalzemeleri=malzemeListesi
    self.malzemeListesiGoster(self.sosMalzemeleri)
    print("hepsini karıştırınız. Karışımı hamurun üzerine sürünüz.\n")
  def icHazirla(self,malzemeListesi):
    self.icMalzemeleri=malzemeListesi
    self.malzemeListesiGoster(self.icMalzemeleri)
    print("hepsini dilimleyip karıştırınız.")
  def firinaSur(self):
    print("Pişirmek için uygun ayarları fırına veriniz ve sonrasında tepsiyi fırına sürünüz.\n")

#Torun Sınıflar
class Pizza(Acik,Firinda):
  def __init__(self):
    super().__init__()
    self.yemekAdi="Pizza"
  def tarifiGoster(self):
    self.hamurMalzemeleri={"tuz":"2 çay kaşığı","un":"2 su bardağı","su":"1 su bardağı"}
    self.sosMalzemeleri={"salça":"2 çorba kaşığı","yağ":"3 çorba kaşığı"}
    self.icMalzemeleri={"domates":"1 adet","kaşar peyniri(rendelenmiş)":"4 kibrit kutusu","mantar":"5 adet","biber":"7 adet","sucuk":"1 kangal"}
    
    print("\nMalzemeler:")
    listele(self.hamurMalzemeleri)
    listele(self.sosMalzemeleri)
    listele(self.icMalzemeleri)

    self.hamuruKar(self.hamurMalzemeleri)
    self.hamuruYogur()
    self.hamuruYay()
    self.sosHazirla(self.sosMalzemeleri)
    self.icHazirla(self.icMalzemeleri)
    print("Karışımı hamurun üzerine dökünüz.\n")
    self.firinaSur()


class Pogaca(Kapali,Firinda):
  def __init__(self):
    super().__init__()
    self.yemekAdi="Poğaça"
  def hamurlariOvalle(self):
    print("Hamur parçalarını poğaçaların kendilerine özgü şeklini verecek şekilde ovalleyiniz.\n")
  def tarifiGoster(self):
    self.hamurMalzemeleri={"tuz":"2 çay kaşığı","un":"2 su bardağı","su":"1 su bardağı"}
    self.sosMalzemeleri={"yumurta":"2 adet","yağ":"3 çorba kaşığı"}
    self.icMalzemeleri={"beyaz peynir":"2 kalıp","maydonoz":"7 dal"}
    
    print("\nMalzemeler:")
    listele(self.hamurMalzemeleri)
    listele(self.sosMalzemeleri)
    listele(self.icMalzemeleri)

    self.hamuruKar(self.hamurMalzemeleri)
    self.hamuruYogur()
    self.hamuruEsitOlarakBol()
    self.hamurlariOvalle()
    self.sosHazirla(self.sosMalzemeleri)
    self.icHazirla(self.icMalzemeleri)
    print("Sonrasında hamurları yan taraflarından yarınız ve karışımı hamurların içlerine eşit miktarda paylaştırarak yerleştiriniz.\n")
    self.firinaSur()

class Pisi(Kapali,Yagda):
  def __init__(self):
    super().__init__()
    self.yemekAdi="Pişi"
  def hamurlaraHalkaSekliVer(self):
    print("Hamur parçalarına pişinin kendine özgü şekli olan halka biçimini veriniz.\n")
  def tarifiGoster(self):
    self.hamurMalzemeleri={"tuz":"2 çay kaşığı","un":"2 su bardağı","su":"1 su bardağı"}
    
    print("\nMalzemeler:")
    listele(self.hamurMalzemeleri)

    self.hamuruKar(self.hamurMalzemeleri)
    self.hamuruYogur()
    self.hamuruEsitOlarakBol()
    self.hamurlaraHalkaSekliVer()
    self.yagdaKizart()
    


#asıl çalışmanın tanımlandığı kısım
def main():
  print("             (: Mutfağımıza Hoş Geldiniz :)      \n")
  print("               Yemek Menümüz aşağıdadır:         \n")
  for index, i in enumerate(yemekListesi):
    print(index+1, i)

  choice=input("\nLütfen tarifini görmek istediğiniz yemeğin adını veya sıra numarasını giriniz. Çıkmak için '0' veya 'çıkış' giriniz:")
  
  while True:
    if choice=="1" or choice=="Pizza":
      pizza=Pizza()
      tarifiGor(pizza)
    elif choice=="2" or choice=="Poğaça":
      pogaca=Pogaca()
      tarifiGor(pogaca)
    elif choice=="3" or choice=="Pişi":
      pisi=Pisi()
      tarifiGor(pisi)
    elif choice=="0" or choice=="çıkış":
      print("\n        .... Yine bekleriz ....")
      break
    else:
      print("Yanlış girişte bulundunuz, lütfen tekrar deneyin!")
      continue

    print("(:  Diğer Yemeklerimizin Enfes Tariflerini de Sizlere Önermekteyiz  :)")
    print("              İncelemeniz için Yemek Menümüz aşağıdadır:         \n")
    for index, i in enumerate(yemekListesi):
      print(index+1, i)

    choice=input("\nLütfen tarifini görmek istediğiniz yemeğin adını veya sıra numarasını giriniz. Çıkmak için '0' veya 'çıkış' giriniz:")
    

#asıl çalışmanın gerçekleştirildiği kısım
main()
```
